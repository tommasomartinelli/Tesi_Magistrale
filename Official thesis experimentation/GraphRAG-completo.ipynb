## Salvataggio grafo sommarizzato per visualizzazione

```python
def save_graphs_with_pyvis(directory):
    if not os.path.exists(directory) or not os.listdir(directory):
        print("empty path")
        return

    for filename in os.listdir(directory):
        if filename.endswith("summarized_graph.graphml"):
            file_path = os.path.join(directory, filename)
            print(f"loading: {file_path}")

            # Load graph using NetworkX
            G = nx.read_graphml(file_path)
            
            net = Network(notebook=False, cdn_resources='in_line')
            
            # add edge and nodes
            net.from_nx(G)
            
            # Save as HTML
            html_filename = "Documents" + os.path.splitext(filename)[0] + ".html"
            net.save_graph(html_filename)
            print(f"Graph saved as: {html_filename}")

save_graphs_with_pyvis(file_path)
```

## Caricamento in Neo4j

```python
def convert_graphml_to_csv(directory):
    if not os.path.exists(directory) or not os.listdir(directory):
        print("The specified path is empty or does not exist.")
        return

    filename = "summarized_graph.graphml"  # Specific file to convert
    file_path = os.path.join(directory, filename)
    
    if not os.path.exists(file_path):
        print(f"The file {filename} does not exist in the specified directory.")
        return

    print(f"Loading: {file_path}")

    G = nx.read_graphml(file_path)
    
    # Extract nodes and edges
    nodes_df = pd.DataFrame(list(G.nodes(data=True)), columns=["id", "attributes"])
    edges_df = pd.DataFrame(list(G.edges(data=True)), columns=["source", "target", "attributes"])

    nodes_df['attributes'] = nodes_df['attributes'].apply(lambda x: str(x))
    edges_df['attributes'] = edges_df['attributes'].apply(lambda x: str(x))
    
    # Save as CSV
    graph_name = os.path.splitext(filename)[0]
    nodes_df.to_csv(f"{graph_name}_nodes.csv", index=False)
    edges_df.to_csv(f"{graph_name}_edges.csv", index=False)
    print(f"Saved {graph_name}_nodes.csv and {graph_name}_edges.csv")

# Example usage
convert_graphml_to_csv(file_path)
```

```python
import ast
from py2neo import Graph, Node, Relationship

def clean_string(s):
    return s.strip().strip('"').strip()

def create_node(row):
    attributes = ast.literal_eval(row['attributes'])
    node_type = clean_string(attributes.get('type', ''))
    entity_type = clean_string(attributes.get('entity_type', node_type))
    name = clean_string(row['id'])
    
    if not node_type:
        node_type = "Unknown"
    if not name:
        name = f"Unnamed_{row.name}" 
    
    node = Node(node_type,
                name=name,
                description=attributes.get('description', ''),
                entity_type=entity_type)
    return node

nodes = {}
for _, row in nodes_df.iterrows():
    node = create_node(row)
    nodes[clean_string(row['id'])] = node
    graph.create(node)

for _, row in edges_df.iterrows():
    source_id = clean_string(row['source'])
    target_id = clean_string(row['target'])
    if source_id in nodes and target_id in nodes:
        source = nodes[source_id]
        target = nodes[target_id]
        attributes = ast.literal_eval(row['attributes'])
        rel = Relationship(source, "RELATED_TO", target, 
                           weight=attributes.get('weight', 1.0), 
                           description=attributes.get('description', ''))
        graph.create(rel)
    else:
        print(f"Skipping relationship: {source_id} -> {target_id} (Node not found)")
```
